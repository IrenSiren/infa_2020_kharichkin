```python
import pygame
from pygame.draw import *
import numpy as np

pygame.init()

FPS = 30

clock = pygame.time.Clock()
```

## Упражнение №1 (заяц)

```python
# names of the colors
BLACK = (0, 0, 0)
YELLOW = (255, 255, 0)
RED = (255, 0, 0)
GREY = (200, 200, 200)

# colors of the parts
screen_color = BLACK
body_color = GREY
head_color = GREY
left_ear_color = GREY
right_ear_color = GREY
left_leg_color = GREY
right_leg_color = GREY

# standart sizes of a hare
body_width = 90
body_height = 200
head_radius = 30
left_ear_width = 15
left_ear_height = 120
right_ear_width = 15
right_ear_height = 120
left_leg_width = 20
left_leg_height = 40
right_leg_width = 20
right_leg_height = 40

# standart arrangement of a hare
head_coord = 0, -130
left_ear_coord = -20, -180
left_ear_angle = 30.0
right_ear_coord = 15, -180
right_ear_angle = 0.0
left_leg_coord = -20, 95
left_leg_angle = 90.0
right_leg_coord = 20, 105
right_leg_angle = -125.0

# Since all body parts of a hare are (inclined) ellipses, this is the function, that draws inclined ellipses.


def draw_inclined_ellipse(surface: pygame.Surface,
                          x: int,
                          y: int,
                          width: int,
                          height: int,
                          theta: float = 0.0,
                          color=BLACK):
    '''
    Функция рисует эллипс, у которого центр соответствует точке (x, y).
    В процессе отрисовки создаётся отдельный объект pygame.Surface, который впоследствии
    накладывается на исходную поверхность surface
    :param surface: объект типа pygame.Surface, на котором производится отрисовка
    :param x, y: координаты центра эллипса
    :param width: минимальная ширина прямоугольника, в который можно вписать эллипс (малая ось эллипса)
    :param height: минимальная высота прямоугольника, в который можно вписать эллипс (большая ось эллипса)
    :param theta: угол между главной осью эллипса и вертикалью
                  (theta = 0 в направлении наверх;
                  theta отсчитывается против часовой стрелки;
                  [theta] = degree;
                  по умолчанию theta=0.0)
    :param color: цвет эллипса (по умолчанию BLACK)
    :return: None
    '''
    subsurface = pygame.Surface((width, height), pygame.SRCALPHA)
    ellipse(subsurface, color, [0, 0, width, height])
    subsurface_rot = pygame.transform.rotate(subsurface, theta)
    # coordinates of top left corner of subsurface_rot    # Создатели pygame.transform.rotate() конченные
    x_rot = min(round(x - np.sqrt((width/2)**2+(height/2)**2)
                      * np.sin(theta*np.pi/180 + np.arctan(width/height))),
                round(x - np.sqrt((width/2)**2+(height/2)**2)
                      * np.sin(theta*np.pi/180 - np.arctan(width/height))),
                round(x + np.sqrt((width/2)**2+(height/2)**2)
                      * np.sin(theta*np.pi/180 + np.arctan(width/height))),
                round(x + np.sqrt((width/2)**2+(height/2)**2)
                      * np.sin(theta*np.pi/180 - np.arctan(width/height))))
    y_rot = min(round(y - np.sqrt((width/2)**2+(height/2)**2)
                      * np.cos(theta*np.pi/180 - np.arctan(width/height))),
                round(y - np.sqrt((width/2)**2+(height/2)**2)
                      * np.cos(theta*np.pi/180 + np.arctan(width/height))),
                round(y + np.sqrt((width/2)**2+(height/2)**2)
                      * np.cos(theta*np.pi/180 - np.arctan(width/height))),
                round(y + np.sqrt((width/2)**2+(height/2)**2)
                      * np.cos(theta*np.pi/180 + np.arctan(width/height))))
    surface.blit(subsurface_rot, [x_rot, y_rot])


def draw_body(surface: pygame.Surface,
              x: int,
              y: int,
              width: int,
              height: int,
              color=BLACK):
    '''
    Функция рисует эллиптическое тело зайца, у которого центр соответствует точке (x, y).
    В процессе отрисовки создаётся отдельный объект pygame.Surface, который впоследствии
    накладывается на исходную поверхность surface
    :param surface: объект типа pygame.Surface, на котором производится отрисовка
    :param x, y: координаты центра эллиптического тела зайца
    :param width: минимальная ширина прямоугольника, в который можно вписать тело зайца (малая ось эллипса)
    :param height: минимальная высота прямоугольника, в который можно вписать тело зайца (большая ось эллипса)
    :param color: цвет тела зайца (по умолчанию BLACK)
    :return: None
    '''
    draw_inclined_ellipse(surface, x, y, width, height, 0.0, color)


def draw_head(surface: pygame.Surface,
              x: int,
              y: int,
              radius: int,
              color=BLACK):
    '''
    Функция рисует круглую голову зайца, у которой центр соответствует точке (x, y).
    В процессе отрисовки создаётся отдельный объект pygame.Surface, который впоследствии
    накладывается на исходную поверхность surface
    :param surface: объект типа pygame.Surface, на котором производится отрисовка
    :param x, y: координаты центра круглой головы зайца
    :param radius: радиус круглой головы зайца
    :param color: цвет головы зайца (по умолчанию BLACK)
    :return: None
    '''
    draw_inclined_ellipse(surface, x, y, 2*radius, 2*radius, 0.0, color)


def draw_ear(surface: pygame.Surface,
             x: int,
             y: int,
             width: int,
             height: int,
             theta: float = 0.0,
             color=BLACK):
    '''
    Функция рисует эллиптическое ухо зайца, у которого центр соответствует точке (x, y), 
    а главная ось наклонена к вертикали под углом theta.
    В процессе отрисовки создаётся отдельный объект pygame.Surface, который впоследствии
    накладывается на исходную поверхность surface
    :param surface: объект типа pygame.Surface, на котором производится отрисовка
    :param x, y: координаты центра эллиптического уха зайца
    :param width: минимальная ширина прямоугольника, в который можно вписать ухо зайца (малая ось эллипса)
    :param height: минимальная высота прямоугольника, в который можно вписать ухо зайца (большая ось эллипса)
    :param theta: угол между главной осью эллиптического уха зайца и вертикалью
                  (theta = 0 в направлении наверх;
                  theta отсчитывается против часовой стрелки;
                  [theta] = degree;
                  по умолчанию theta=0.0)
    :param color: цвет уха зайца (по умолчанию BLACK)
    :return: None
    '''
    draw_inclined_ellipse(surface, x, y, width, height, theta, color)


def draw_leg(surface: int,
             x: int,
             y: int,
             width: int,
             height: int,
             theta: float = 0.0,
             color=BLACK):
    '''
     Функция рисует эллиптическое ухо зайца, у которого центр соответствует точке (x, y), 
     а главная ось наклонена к вертикали под углом theta.
     В процессе отрисовки создаётся отдельный объект pygame.Surface, который впоследствии
     накладывается на исходную поверхность surface
     :param surface: объект типа pygame.Surface, на котором производится отрисовка
     :param x, y: координаты центра эллиптической ноги зайца
     :param width: минимальная ширина прямоугольника, в который можно вписать ногу зайца (малая ось эллипса)
     :param height: минимальная высота прямоугольника, в который можно вписать ногу зайца (большая ось эллипса)
     :param theta: угол между главной осью эллиптической ноги зайца и вертикалью
                   (theta = 0 в направлении наверх;
                   theta отсчитывается против часовой стрелки;
                   [theta] = degree;
                   по умолчанию theta=0.0)
     :param color: цвет ноги зайца (по умолчанию BLACK)
     :return: None
     '''
    draw_inclined_ellipse(surface, x, y, width, height, theta, color)


def draw_hare(surface: pygame.Surface,
              x: int,
              y: int,
              size: float = 1.0):
    '''
    Функция рисует зайца, у которого центр эллиптического тела соответствует точке (x, y)
    :param surface: объект типа pygame.Surface, на котором производится отрисовка
    :param x, y: координаты центра эллиптического тела
    :param size: отношение размеров зайца к стандартным размерам зайца
    :return: None
    '''
    # drawing body
    draw_body(surface, x, y, round(body_width * size),
              round(body_height * size), body_color)

    # drawing head
    draw_head(surface, x + round(head_coord[0] * size), y + round(
        head_coord[1] * size), round(head_radius * size), head_color)

    # drawing left ear
    draw_ear(surface, x + round(left_ear_coord[0] * size), y + round(left_ear_coord[1] * size),
             round(left_ear_width * size),
             round(left_ear_height * size), left_ear_angle, left_ear_color)

    # drawing right ear
    draw_ear(surface, x + round(right_ear_coord[0] * size), y + round(right_ear_coord[1] * size),
             round(right_ear_width * size),
             round(right_ear_height * size), right_ear_angle, right_ear_color)

    # drawing left leg
    draw_leg(surface, x + round(left_leg_coord[0] * size), y + round(left_leg_coord[1] * size),
             round(left_leg_width * size),
             round(left_leg_height * size), left_leg_angle, left_leg_color)

    # drawing right leg
    draw_leg(surface, x + round(right_leg_coord[0] * size), y + round(right_leg_coord[1] * size),
             round(right_leg_width * size),
             round(right_leg_height * size), right_leg_angle, right_leg_color)


screen = pygame.display.set_mode((400, 400))
screen.fill(screen_color)
draw_hare(screen, 200, 250, 1.0)
```

## Не забывай запускать эту ячейку

```python
pygame.display.update()
finished = False

while not finished:
    clock.tick(FPS)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            finished = True

pygame.quit()
```
